```python
def convert(p):
  p = str(p)
  p = p.split()
  p = p[0]
  return p
```

```python
data['Power'] = data['Power'].apply(convert)
```

```python
data['Power'].unique()
```

```python
def filter(p):
  if p.isdigit():
    return float(p)
  else:
    return 1.0
```

```python
data['Power'] = data['Power'].apply(filter)
```

```python
data['Power'] = data['Power'].replace(1.0, data['Power'].mean())
```

```python
data = pd.read_csv('/content/Sleep_health_and_lifestyle_dataset.csv')
data.sample(10)
```

```python
def numb(c):
  if c == 'Normal':
    return 2
  elif c == 'Normal Weight':
    return 2
  elif c == 'Overweight':
    return 1
  elif c == 'Obese':
    return 0
  else:
    return 0
```

```python
data['BMI Category'] = data['BMI Category'].apply(numb)
```

```python
data[['High Pressure', 'Low Pressure']] = data['Blood Pressure'].str.split('/', expand=True)
```

```python
def converttype(n):
  n = int(n)
  return n
```

```python
data['High Pressure'] = data['High Pressure'].apply(converttype)
```

```python
data['Low Pressure'] = data['Low Pressure'].apply(converttype)
```

```python
data = data.drop('Blood Pressure', axis=1)
```

```python
data.head()
```

```python
data.head()
```

```python
data = data.drop(['Person ID' , 'Sleep Duration','Stress Level','Physical Activity Level'], axis = 1)
```

```python
data.head()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
def transform(m):
    if(m=='None'):
      return 0
    elif(m=='Sleep Apnea'):
      return 1
    elif(m=='Insomnia'):
      return 2
```

```python
data['Sleep Disorder'] = data['Sleep Disorder'].apply(transform)
```

```python
data = data.drop(['Gender'], axis = 1)
```

```python
data.head()
```

```python
y = data['Sleep Disorder']
```

```python
y
```

```python
x = data.drop(['Sleep Disorder'], axis = 1)
```

```python
x = pd.get_dummies(x, drop_first=True)
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

# AI

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
hyperparameters = {'kernel':('linear', 'rbf'), 'C':[1, 10]}
model = SVC()
clf = GridSearchCV(model, hyperparameters)
clf.fit(X_train, y_train)
```
